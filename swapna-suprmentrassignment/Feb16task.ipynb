```python
import pandas as pd
import numpy as np
data={
    'Age':[23,22,None,25,None],
    'Salary':[50000,None,45000,None,48000],
    'City':['Hyderabad','HYDERABAD','Mumbai','hyderabad','Pune'],
    'Experience':[5,2,3,5,3]        
}
df=pd.DataFrame(data)
print(df)
```

```python
df.duplicated()
df.drop_duplicates()
```

```python
df.isnull()
```

```python
df.dropna()
df.dropna(axis=1)
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].median(),inplace=True)
df['Experience'].fillna(df['Experience'].mode()[0],inplace=True)
print(df)
```

```python
df['City']=df['City'].str.upper()
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[['Age','Salary']]=scaler.fit_transform(df[['Age','Salary']])
print("Final Cleaned Data:\n")
print(df)
```
