```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data={
    'War':['World war I','World war II'],
    'Duration':[4,6],
    'Year':[1914,1939]
}
df=pd.DataFrame(data)
print(df)
```

```python
plt.bar(df['War'],df['Duration'],color='Blue')
plt.title("World war Years")
plt.xlabel("War")
plt.ylabel("Duration")
plt.show()
```

```python
plt.pie(df['Duration'],labels=df['War'],autopct='%1.1f%%',startangle=140)
plt.title('Distribution of Total World war years')
plt.show()
```

```python
plt.hist(df['Year'],bins=[1910,1920,1930,1940,1950],color='purple',edgecolor='black')
plt.title("Frequency of War Start Dates Decades")
plt.xlabel("Year")
plt.ylabel("Number of wars Started")
plt.show()
```
