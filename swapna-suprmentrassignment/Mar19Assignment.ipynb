Mini Project Proposal

Title:
Smart Text Generator using NLP

Problem:
Generating meaningful text from small input is difficult. Basic models produce repetitive output.

Dataset:
Custom text dataset (messy sentences with emojis, slang, typos)

Algorithm:
Markov Chain (Word Prediction)

Methodology:
Clean text
Build word relationships
Predict next word
Generate sentence

Expected Output:
10-word meaningful sentence
Reduced repetition
Better prediction
