```python
from gensim.models import Word2Vec
sentences=[
     ["I", "Love","food"],
     ["I", "Like","Pani puri"],
     ["Biriyani","is","a","famous","in","SouthIndia"],
     ["Dosa","is","famous","food"],
     ["Fishes","would","be","decilious","food","near","seashore"],
     ["Biriyani","is","famous","dish","in","Hyderabad"],
     ["PavBhaji","is","famous","chat","in","Maharastra"],
     ["Rajmachawal","is","famous","in","NorthIndia"],
     ["Poratta","is","one","of","the","tastiest","food","in","Kerala"],
     ["Readymade","porattas","is","available","everywhere"],
     ["Idly","vada","is","well","known","breakfast"],
     ["Chicken","Biriyani","is","my","Favourite"],
     ["MysorePak","is","bestest","sweet"],
     ["Coffee","is","main","species","in","Kodagu"],
     ["I","hate","Pizza"]
    ]
model=Word2Vec(sentences,vector_size=50,window=3,min_count=1,workers=4)
print("Vector for 'food':")
print(model.wv["food"])
print("\n5 similar Words to 'famous':")
print(model.wv.most_similar("famous",topn=5))
```
