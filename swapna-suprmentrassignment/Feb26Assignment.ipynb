```python
import pandas as pd
import numpy as np
data={
    'Name':['navya','arjun','riya','arjun',None],
    'Age':[23,25,None,25,22],
    'Salary':[50000,None,45000,60000,48000],
    'Department':['HR','IT','it','HR','it']
}
df=pd.DataFrame(data)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna()
df.dropna(axis=1)
```

```python
df['Name'].fillna('Unknown',inplace=True)
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].median(),inplace=True)
df['Department'].fillna(df['Department'].mode(),inplace=True)
print(df)
```

```python
df.duplicated()
df.drop_duplicates()
```

```python
df['Department']=df['Department'].str.upper()
print(df)
```

```python
print("Cleaned Dataset:")
print(df)
```

Data Cleaning is a process to remove error,fixing inconsistency,standardizing formats,preparing raw data for analysis
