```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents=[
    "I Love NLP",
    "I like Artificial Intelligence",
    "I love Cats",
    "Coding is fun",
    "Dogs are kindest"
]
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(documents)
words=vectorizer.get_feature_names_out()
for i in range(len(documents)):
    print("\nDocument:",documents[i])
    scores=X[i].toarray()[0]
    for j in scores.argsort()[-3:][::-1]:
        print(words[j],":",round(scores[j],3))
```

Explanation: TF-IDF is a term frequency-Inverse document frequency which is used to find the important words in a document and in the output- are,artificial,cats,coding,dogs,fun,intelligence,is,kindest,like,love,nlp are identified as top keywords because these are the words that appear frequently in the given document
