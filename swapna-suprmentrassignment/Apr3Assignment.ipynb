```python
from sklearn.feature_extraction.text import CountVectorizer

class SimpleChatbot:
    def respond(self, text):
        text = text.lower()

        if "hello" in text:
            return "Hi! How can I help you?"
        elif "ai" in text:
            return "AI stands for Artificial Intelligence."
        elif "machine learning" in text:
            return "Machine Learning is a subset of AI that learns from data."
        else:
            return "I am still learning!"

def extract_keywords(text):
    temp_words = text.lower().split()
    
    stopwords = {"who", "are", "you", "is", "am", "the", "a", "an", "to", "me", "about"}
    filtered = [w for w in temp_words if w not in stopwords]

    if len(filtered) == 0:
        return ["no_keywords_found"]

    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform([" ".join(filtered)])

    return vectorizer.get_feature_names_out()

bot = SimpleChatbot()

msg = input("You: ")

print("Chatbot:", bot.respond(msg))
print("Keywords:", list(extract_keywords(msg)))
```
