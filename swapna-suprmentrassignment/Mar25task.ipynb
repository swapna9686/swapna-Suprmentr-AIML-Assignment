```python
from collections import Counter
text=input("Enter the sentence:")
words=text.split()
model={}
for i in range(len(words)-1):
    word=words[i]
    next_word=words[i+1]
    if word not in model:
        model[word]=[]
    model[word].append(next_word)
def predict_smart(word):
    if word in model:
        return Counter(model[word]).most_common(1)[0][0]
    else:
        return "unknown"
def generate_text(start_word,length=10):
    result=[start_word]
    for _ in range(length-1):
        next_word=predict_smart(result[-1])
        if next_word!="unknown":
            result.append(next_word)
        else:
            break
    return " ".join(result)
start=input("Enter the starting word:")
print(generate_text(start,10))
```
