```python
from gensim.models import Word2Vec
sentences=[
    ["king","queen","man","woman"],
    ["car","bus","train","vehicle"],
    ["apple","banana","fruit"],
    ["dog","cat","animal"],
    ["teacher","student","school"]
]
model=Word2Vec(sentences,vector_size=50,window=2,min_count=1,workers=1)
pairs=[
    ("king","queen"),
    ("car","bus"),
    ("apple","banana"),
    ("dog","cat"),
    ("teacher","student")
]
for w1,w2 in pairs:
    similarity=model.wv.similarity(w1,w2)
    print(w1,"-",w2,":",round(similarity,2))
```

Word2Vec is a technique that learns word embeddings by predicting realationships between words.Word2Vec is used to find the semantic realtionship between words.Words in similar contexts have higher similarity scores.The values range from 0 to 1,where higher values indicate stronger similarity
