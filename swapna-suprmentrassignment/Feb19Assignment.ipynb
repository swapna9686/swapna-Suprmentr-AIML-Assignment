```python
Marks={"Priya":90,"Karthik":85,"Swapna":93,"Prajna":82,"Roja":78}
print("Student Marks:",Marks)
```

```python
topper=max(Marks,key=Marks.get)
print("Topper:",topper,"-",Marks[topper])
```

```python
avg=sum(Marks.values())/len(Marks)
print("Average:",avg)
```

```python
grades={}
for student,marks in Marks.items():
    if marks>=90:
        grades[student]="A"
    elif marks>=80:
        grades[student]="B"
    elif marks>=70:
        grades[student]="C"
    else:
        grades[student]="D"
print("Grades:",grades)
```
