```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("female.png") 
img_array = np.array(img)

print("Shape:", img_array.shape)
print("Pixels:\n", img_array)
print("Channels:", img_array.shape[2] if len(img_array.shape) == 3 else 1)

plt.imshow(img_array)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

🔹An image is stored as a matrix of numbers called pixels.

🔹Each pixel contains RGB values (Red, Green, Blue).

🔹Each RGB value ranges from 0 to 255, showing color intensity.

🔹Image shape represents height, width, and number of channels (3 for color images).

🔹Therefore, images are numerical data that represent visual information.
