```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data={
    "Age":[22,23,24,45,46,47,30,31,32],
    "Annual_Income":[50,60,80,90,88,95,100,65,62],
    "Spending_score":[5,6,3,9,13,20,20,12,15]
}
df=pd.DataFrame(data)
print(df)
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
labels=kmeans.labels_
print(labels)
```

```python
centroids=kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df['Annual_Income'],df['Spending_score'],c=labels)
plt.scatter(centroids[:,1],centroids[:,2],marker='X',s=200)
plt.title("Customer Segmentation")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
